```python
import pandas as pd
from gmplot import gmplot

# Load the dataset from the Excel file
file_path = 'D:/Semester3/Maysem/Project/Plot_CFB_UsingGM/final_engineered_data_tft.csv'
df = pd.read_csv(file_path)

```

```python
df
```

```python
# Assuming your dataset has columns 'latitude', 'longitude', and 'value' (CFB)
latitudes = df['lat']
longitudes = df['lon']
cfb_values = df['cfb']

```

```python
# Filter CFB values greater than 1
# Filter the CFB values greater than 1
filtered_df = df[df['cfb'] > 1]
```

```python
grouped_df = filtered_df.loc[filtered_df.groupby('locality')['cfb'].idxmax()]
```

```python
# Extract the latitude, longitude, and CFB values
latitudes = grouped_df['lat']
longitudes = grouped_df['lon']
cfb_values = grouped_df['cfb']
```

```python

# Create a gmplot instance
gmap = gmplot.GoogleMapPlotter(latitudes.mean(), longitudes.mean(), 6)
```

```python
# Plot the heatmap
gmap.heatmap(latitudes, longitudes, weights=cfb_values, radius=20, dissipating=True, max_intensity=max(cfb_values), opacity=0.6)
```

```python
# Save the map to an HTML file
gmap.draw('wildfire_prediction_heatmap.html')
```

```python
print("Heatmap has been generated and saved as 'wildfire_prediction_heatmap.html'")
```

```python
# Extract the CSV file used for the plot
output_csv_path = 'D:/Semester3/Maysem/Project/Plot_CFB_UsingGM/locality_Filter_data.csv'
grouped_df.to_csv(output_csv_path, index=False)
print(f"CSV file with plot data has been saved as '{output_csv_path}'")
```

```python
import json
# Create a JSON file with the same data
json_data = grouped_df.to_dict(orient='records')
output_json_path = 'D:/Semester3/Maysem/Project/Plot_CFB_UsingGM/locality_filter_data.json'
with open(output_json_path, 'w') as json_file:
    json.dump(json_data, json_file, indent=2)
print(f"JSON file with plot data has been saved as '{output_json_path}'")
```
