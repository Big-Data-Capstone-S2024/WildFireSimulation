#### Create a vector counting occurrences of each emotion label

```python
from collections import Counter
from scipy.sparse import csr_matrix

def create_book_emotion_vectors(df):
    all_emotions = df['emotions'].unique()
    # Function to create emotion vector for a book
    def create_emotion_vector(book_emotions):
        emotion_counts = Counter(book_emotions)
        vector = np.zeros(len(all_emotions), dtype=int)
        for emotion, count in emotion_counts.items():
            vector[emotion] = count
        return vector

    # Group by book_id and create emotion vectors
    book_emotion_vectors = df.groupby('book_id')['emotions'].apply(create_emotion_vector)
    
    # Convert to a DataFrame for easier handling
    book_emotion_df = pd.DataFrame(book_emotion_vectors.tolist(), index=book_emotion_vectors.index, columns=all_emotions)
    
    # Normalize the vectors
    book_emotion_df = book_emotion_df.div(book_emotion_df.sum(axis=1), axis=0)
    
    # Convert to CSR matrix for efficiency
    book_emotion_csr = csr_matrix(book_emotion_df.values)
    
    return book_emotion_csr

book_emotion_matrix = create_book_emotion_vectors(merged_df)
```

#### Averaging reviews embedded per book

```python
from scipy.sparse import csr_matrix, save_npz

# Define the conversion function
def convert_to_array(x):
    if isinstance(x, str):
        # Remove newline characters and extra spaces
        cleaned_str = x.replace('\n', ' ')
        return np.fromstring(cleaned_str.strip("[]"), sep=' ')
    elif isinstance(x, (list, np.ndarray)):
        return np.array(x)
    else:
        return np.array([])  # Return an empty array for invalid entries

def create_book_avg_embeddings(df):
    # Convert 'reviews_embedding' to numpy arrays for easier computation
    df['reviews_embedding'] = df['reviews_embedding'].apply(convert_to_array)
    
    # Filter out invalid entries where the embedding array is empty
    df = df[df['reviews_embedding'].apply(len) > 0]

    # Group by 'book_id' and compute average embedding
    average_embeddings = df.groupby('book_id')['reviews_embedding'].apply(lambda x: np.mean(np.vstack(x), axis=0))
    # Convert to a DataFrame for easier handling
    book_avg_embeddings_df = pd.DataFrame(average_embeddings.tolist(), index=average_embeddings.index)
    
    # Convert to CSR matrix for efficiency
    book_avg_embeddings_csr = csr_matrix(book_avg_embeddings_df.values)
    
    return book_avg_embeddings_csr

book_avg_embeddings_matrix = create_book_avg_embeddings(merged_df)
```

```python
# Create mappings for user_id and book_id to matrix indices
user_ids = merged_df['user_id'].unique()
book_ids = merged_df['book_id'].unique()
user_id_map = {id: i for i, id in enumerate(user_ids)}
book_id_map = {id: i for i, id in enumerate(book_ids)}
print('number of users: ', len(user_ids))
print('number of books: ', len(book_ids))
```

```python
from sklearn.preprocessing import MinMaxScaler

def create_user_book_csr_matrix(df):
    # Create row and column indices
    row_indices = [user_id_map[id] for id in df['user_id']]
    col_indices = [book_id_map[id] for id in df['book_id']]
    
    # Normalize the review scores
    min_score = df['review_score'].min()
    max_score = df['review_score'].max()
    normalized_scores = (df['review_score'] - min_score) / (max_score - min_score)
    
    # Create the CSR matrix
    csr_matrix_ub = csr_matrix((normalized_scores, (row_indices, col_indices)), 
                               shape=(len(user_ids), len(book_ids)))
    
    return csr_matrix_ub

user_book_matrix = create_user_book_csr_matrix(merged_df)
```
